Process US Average Temperature (Fahrenheit) Data 1970-2022

```python
import pandas as pd
import os
import datetime
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

current_dir = os.getcwd()
```

```python
path = os.path.join(current_dir, '..', 'Data', 'US_avg_temp_data_1966-2022.csv')

temp_df = pd.read_csv(path, skiprows=4)
temp_df = temp_df.rename(columns={'Value': 'Temperature'})
temp_df.head()

# Datetime objects as continuous variable?

#for index, row in temp_df.iterrows():
    #temp_df.at['Date']
```

```python
# Visualize Average Temperature over the years/months
plt.figure()
plt.title('Average Temperature (Fahrenheit)')
plt.xlabel('Date')
plt.ylabel('Average Temperature (Fahrenheit)')
plt.title('US Average Temperature Trends 1970-2022')
plt.plot(temp_df['Date'], temp_df['Temperature'])
```

```python
# Regression Line on Overall Trend
dates = np.reshape(np.array(temp_df['Date']), (-1, 1))
temps = np.reshape(np.array(temp_df['Temperature']), (-1, 1))

reg = LinearRegression().fit(dates, temps)
print(reg.coef_)
print(reg.intercept_)

plt.figure()
plt.title('Average Temperature (Fahrenheit)')
plt.xlabel('Date')
plt.ylabel('Average Temperature (Fahrenheit)')
plt.title('US Average Temperature Trends 1970-2022')
plt.plot(temp_df['Date'], temp_df['Temperature'])
plt.plot(temp_df['Date'], reg.intercept_ + temp_df['Date']*reg.coef_[0, 0])
```
